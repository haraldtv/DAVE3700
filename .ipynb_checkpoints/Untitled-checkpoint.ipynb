```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 500
XMIN = -2*np.pi
XMAX = 2*np.pi

#ax = plt.figure(figsize =(3, 2), dpi=400).add_subplot(projection='3d')
ax = plt.figure().add_subplot(projection='3d')

# Prepare arrays x, y, z
z = np.linspace(XMIN, XMAX, N)

x = z * np.sin(2*z)
y = z * np.cos(2*z)

ax.plot(x, y, z, label="DAVE3700")
ax.legend()
```

```python
N = 4
XMIN = -1
XMAX = 1

ax = plt.figure().add_subplot(projection='3d')

# Make the grid
x, y, z = np.meshgrid(np.linspace(XMIN, XMAX, N),
                      np.linspace(XMIN, XMAX, N),
                      np.linspace(XMIN, XMAX, N))

# Make the direction data for the arrows
u = -y
v = x
w = 0

ax.quiver(x, y, z, u, v, w, length=0.6, normalize=False)
```
